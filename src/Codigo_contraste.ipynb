```python
import os
from PIL import Image, ImageEnhance
```

```python
directory = '/workspaces/LungsXray-FP/data/raw/NORMAL'
dir_target = '/workspaces/LungsXray-FP/data/raw/Imagenes_contraste/NORMAL'
name = "NORMAL"

num = 0
for filename in os.listdir(directory):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        filename1 = os.path.join(directory, filename)
        im = Image.open(filename1)

        # Convert image to RGB mode if not already in RGB
        if im.mode != "RGB":
            im = im.convert("RGB")

        # Enhance brightness and contrast
        im = ImageEnhance.Brightness(im).enhance(0.7)
        im = ImageEnhance.Contrast(im).enhance(2.0)

        # Resize the image
        im.thumbnail((299, 299), Image.BICUBIC)  # Use BICUBIC instead of ANTIALIAS
        
        num += 1
        if num == 100000:
            break
        
        num_str = str(num)
        file_tar = os.path.join(dir_target, name + "-" + num_str + ".png")
        
        im.save(file_tar)
```
